# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from newsapi import NewsApiClient
from dotenv import load_dotenv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
load_dotenv()


%matplotlib inline
```

```python
# Read your api key environment variable
api_key = os.getenv('NEWS_API_KEY')
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_news = newsapi.get_everything(q='Bitcoin', language='en')
bitcoin_news
```

```python
# Fetch the Ethereum news articles
ethereum_news = newsapi.get_everything(q='Ethereum', language='en')
ethereum_news
```

```python
# Create the Bitcoin sentiment scores DataFrame

def sentiment_scores_dataframe(news_headlines):

    news_sentiments = []

    for article in news_headlines["articles"]:
        try:
            text = article["content"]
            date = article["publishedAt"][:10]
            sentiment = analyzer.polarity_scores(text)
            compound = sentiment["compound"]
            pos = sentiment["pos"]
            neu = sentiment["neu"]
            neg = sentiment["neg"]
        
            news_sentiments.append({
                "text": text,
                "date": date,
                "compound": compound,
                "positive": pos,
                "negative": neg,
                "neutral": neu
            
            })
        
        except AttributeError:
            pass
        
    # Create DataFrame
    news_sentiment_df = pd.DataFrame(news_sentiments)
    
    # Reorder DataFrame columns
    cols = ["date", "text", "compound", "positive", "negative", "neutral"]
    news_sentiment_df = news_sentiment_df[cols]

    # Return DataFrame
    return news_sentiment_df 

bitcoin_df = sentiment_scores_dataframe(bitcoin_news)
bitcoin_df.head()

```

```python
# Create the ethereum sentiment scores DataFrame
ethereum_df = sentiment_scores_dataframe(ethereum_news)
ethereum_df.head()
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

### Questions:

Q: Which coin had the highest mean positive score?

A: Bitcoin

Q: Which coin had the highest compound score?

A: Ethereum has the highest max compound score.

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary
sw_add_ons = {'','chars','th'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Create a list of the words
    words_tokenized = word_tokenize(text)


    # Convert the words to lowercase
    lower = [word.lower() for word in words_tokenized]
    
    # Remove the punctuation
    regex = re.compile('[^a-zA-Z]')
    words_regexed = [regex.sub('', word) for word in lower]
    
    # Remove the stop words
    sw = set(stopwords.words('english'))
    sw_removed = [word.lower() for word in words_regexed if word not in sw.union(sw_add_ons)]
    # Lemmatize Words into root words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in sw_removed]
    
    return tokens
```

```python
# Create a new tokens column for bitcoin
bitcoin_tokens = [tokenizer(headline) for headline in bitcoin_df['text']]
bitcoin_df['Tokens'] = bitcoin_tokens
bitcoin_df.head()
```

```python
# Create a new tokens column for ethereum
ethereum_tokens = [tokenizer(headline) for headline in ethereum_df['text']]
ethereum_df['Tokens'] = ethereum_tokens
ethereum_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
def join_column_to_combined_string(df_column):
    '''Joins dataframe column of strings and returns one big string.'''
    combined_string = [' '.join(item) for item in df_column]
    return combined_string

# Created custom function to get bigrams
def get_bigrams(combined_string):
    # Run tokenizer to clean data
    processed = tokenizer(str(combined_string))

    # Create bigrams list with N=2
    bigrams = ngrams(processed, n=2)

    # Return bigrams
    return bigrams

# Join btc dataframe columns to string
bitcoin_combined_string = join_column_to_combined_string(bitcoin_df['Tokens'])

# Create btc bigrams
bitcoin_bigrams = get_bigrams(bitcoin_combined_string)
# Get the top 10 bigrams to show
bitcoin_count = Counter(bitcoin_bigrams).most_common(10)
bitcoin_count
```

```python
# Generate the Ethereum N-grams where N=2
# Join dataframe columns to string
ethereum_combined_string = join_column_to_combined_string(ethereum_df['Tokens'])
# Create bigrams
ethereum_bigrams = get_bigrams(ethereum_combined_string)
# Get the top 10 bigrams to show
ethereum_count = Counter(ethereum_bigrams).most_common(20)
ethereum_count
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
bitcoin_combined_string = ' '.join(bitcoin_df['text'])
bitcoin_words_tokenized = tokenizer(bitcoin_combined_string)
bitcoin_word_count = token_count(bitcoin_words_tokenized)
bitcoin_word_count
```

```python
# Get the top 10 words for Ethereum
ethereum_combined_string = ' '.join(ethereum_df['text'])
ethereum_words_tokenized = tokenizer(ethereum_combined_string)
ethereum_word_count = token_count(ethereum_words_tokenized)
ethereum_word_count
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
bitcoin_combined_string
```

```python
# Run the NER processor on all of the text
bitcoin_ner = nlp(bitcoin_combined_string)

# Add a title to the document
bitcoin_ner.user_data['title'] = 'Bitcoin NER'
```

```python
# Render the visualization
displacy.render(bitcoin_ner, style='ent')
```

```python
# List all Entities
for entity in bitcoin_ner.ents:
    print(entity.text, entity.label_)
```
